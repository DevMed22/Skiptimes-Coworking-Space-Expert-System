```python
from experta import *
```

```python
class CoworkingSpace(KnowledgeEngine):
    @DefFacts()
    def _intialize(self):
        print("Hello now you can book a Course or Room \n let's try")
        yield Fact(action = "find_book")
            
    @Rule(NOT(Fact(Course = W())))
    def value1(self):
        self.declare(Fact(Course = input('Do you want to book a Course?')))
    
    @Rule(NOT(Fact(Rooms = W())))
    def value2(self):
        self.declare(Fact(Rooms = input('Do you want book a Room?')))
    ###############################################################################################
    @Rule(Fact(Course = 'Yes'), NOT(Fact(Programming = W())))
    def value3(self):
        self.declare(Fact(Programming = input('Do you want programming course?')))
        
    @Rule(Fact(Course = 'Yes'), NOT(Fact(Design = W())))
    def value4(self):
        self.declare(Fact(Design = input('Do you want Design course?')))
    
    @Rule(Fact(Course = 'Yes'), NOT(Fact(Language = W())))
    def value5(self):
        self.declare(Fact(Language = input('Do you want Language course?')))
    ###############################################################################################
    @Rule(Fact(Course = 'Yes'), Fact(Programming = 'Yes'), NOT(Fact(Web = W())))
    def value6(self):
        self.declare(Fact(Web = input('Do you want Web course?')))
    
    @Rule(Fact(Course = 'Yes'), Fact(Programming = 'Yes'), NOT(Fact(Mobile = W())))
    def value7(self):
        self.declare(Fact(Mobile = input('Do you want Mobile course?')))
        
    @Rule(Fact(Course = 'Yes'), Fact(Programming = 'Yes'), NOT(Fact(Data_Science = W())))
    def value8(self):
        self.declare(Fact(Data_Science = input('Do you want Data Science course?')))
        
    ##############################################################################################
    @Rule(Fact(Course = 'Yes'), Fact(Programming = 'Yes'), Fact(Web = 'Yes'), NOT(Fact(Frontend = W())))
    def value9(self):
        self.declare(Fact(Frontend = input('Do you want Frontend course?')))
        
    @Rule(Fact(Course = 'Yes'), Fact(Programming = 'Yes'), Fact(Web = 'Yes'), NOT(Fact(Backend = W())))
    def value10(self):
        self.declare(Fact(Backend = input('Do you want Backend course?')))
    ########
    @Rule(Fact(Course = 'Yes'), Fact(Programming = 'Yes'), Fact(Web = 'Yes'), (Fact(Frontend = 'Yes')))
    def value11(self):
        print("**Great choice, In frontend course, you will learn React framwork to build the awesome websites**")
    
    @Rule(Fact(Course = 'Yes'), Fact(Programming = 'Yes'), Fact(Web = 'Yes'), (Fact(Backend = 'Yes')))
    def value12(self):
        print("**Awesome, Through Learning Backend with Django, you will be able to create great Websites in no time**")
        
    #######################################################################################################################
    @Rule(Fact(Course = 'Yes'), Fact(Programming = 'Yes'), Fact(Mobile = 'Yes'), Fact(Cross_platform = 'Yes'))
    def value13(self):
        print("**Flutter will enable you to build cross paltform application that work everywhere**")
    
    @Rule(Fact(Course = 'Yes'), Fact(Programming = 'Yes'), Fact(Mobile = 'Yes'), NOT(Fact(Native = W())))
    def value14(self):
        self.declare(Fact(Native = input('Do you want Native course?')))
          
    #################################################################################################################
    @Rule(Fact(Course = 'Yes'), Fact(Programming = 'Yes'), Fact(Mobile = 'Yes'), Fact(Native = 'Yes'), NOT(Fact(Andriod = W())))
    def value15(self):
        self.declare(Fact(Andriod = input('Do you want Andriod course?')))

    @Rule(Fact(Course = 'Yes'), Fact(Programming = 'Yes'), Fact(Mobile = 'Yes'), Fact(Native = 'Yes'), NOT(Fact(IOS = W())))
    def value16(self):
        self.declare(Fact(IOS = input('Do you want IOS course?')))
    #############
    @Rule(Fact(Course = 'Yes'), Fact(Programming = 'Yes'), Fact(Mobile = 'Yes'), Fact(Native = 'Yes'), Fact(Andriod = 'Yes'))
    def value17(self):
        print("**Use Kotlin to build Andriod apps that will be used by millions of people**") 
          
    @Rule(Fact(Course = 'Yes'), Fact(Programming = 'Yes'), Fact(Mobile = 'Yes'), Fact(Native = 'Yes'), Fact(IOS = 'Yes'))
    def value18(self):
        print("**Use Swift to build IOS apps that will be used by iPhone users**")
    ####################################################################################################################
    @Rule(Fact(Course = 'Yes'), Fact(Programming = 'Yes'), Fact(Data_Science = 'Yes'), NOT(Fact(ML = W())))
    def value19(self):
        self.declare(Fact(ML = input('Do you want Machine Learning Course?')))
    
    @Rule(Fact(Course = 'Yes'), Fact(Programming = 'Yes'), Fact(Data_Science = 'Yes'), NOT(Fact(Data_Analysis = W())))
    def value20(self):
        self.declare(Fact(Data_Analysis = input('Do you want Data Analysis Course?')))
          ############
    @Rule(Fact(Course = 'Yes'), Fact(Programming = 'Yes'), Fact(Data_Science = 'Yes'), (Fact(ML = 'Yes')))
    def value21(self):
        print("**Build Machine Learning Models using TensorFlow and Pytorch**")
    @Rule(Fact(Course = 'Yes'), Fact(Programming = 'Yes'), Fact(Data_Science = 'Yes'), (Fact(Data_Analysis = 'Yes')))
    def value22(self):
        print("**Show Data insights using Python, SQL, Excel and Power BI**")
    #################################################################################################################
    @Rule(Fact(Course = 'Yes'), Fact(Design = 'Yes'), NOT(Fact(UI = W())))
    def value23(self):
          self.declare(Fact(UI = input('Do you want UI Course?')))
    
    @Rule(Fact(Course = 'Yes'), Fact(Design = 'Yes'), NOT(Fact(Graphic_Design = W())))
    def value24(self):
          self.declare(Fact(Graphic_Design = input('Do you want Graphic Design Course?')))
          
    @Rule(Fact(Course = 'Yes'), Fact(Design = 'Yes'), NOT(Fact(Motion_Graphics = W())))
    def value25(self):
          self.declare(Fact(Motion_Graphics = input('Do you want Motion Graphics Course?')))
          
    ###########################################################################################################
    
    @Rule(Fact(Course = 'Yes'), Fact(Design = 'Yes'), Fact(UI = 'Yes'))
    def value26(self):
          print("**Using Figma you will create great UI designs**")
                
    @Rule(Fact(Course = 'Yes'), Fact(Design = 'Yes'), Fact(Graphic_Design = 'Yes'))
    def value27(self):
        print("**You will use adobe products like Photoshop and Illistrator to create Graphic Designs**")
          
    @Rule(Fact(Course = 'Yes'), Fact(Design = 'Yes'), Fact(Motion_Graphics = 'Yes'))
    def value28(self):
        print("**Use After Effects to create motion graphics**")
          
    ##########################################################################################################
          
    @Rule(Fact(Course = 'Yes'), Fact(Language = 'Yes'), NOT(Fact(English = W())))
    def value29(self):
          self.declare(Fact(English = input('Do you want English Course?')))
    
    @Rule(Fact(Course = 'Yes'), Fact(Language = 'Yes'), NOT(Fact(German = W())))
    def value30(self):
          self.declare(Fact(German = input('Do you want German Course?')))
          
    @Rule(Fact(Course = 'Yes'), Fact(Language = 'Yes'), NOT(Fact(Spanish = W())))
    def value31(self):
          self.declare(Fact(Spanish = input('Do you want Spanish Course?')))
          
    #########################################################################################################
    
    @Rule(Fact(Course = 'Yes'), Fact(Language = 'Yes'), Fact(English = 'Yes'))
    def value32(self):
        print("**From Beginner to Advanced in English in no time**")
                
    @Rule(Fact(Course = 'Yes'), Fact(Language = 'Yes'), Fact(German = 'Yes'))
    def value33(self):
        print("**Best Education is in Germany**")
          
    @Rule(Fact(Course = 'Yes'), Fact(Language = 'Yes'), Fact(Spanish = 'Yes'))
    def value34(self):
        print("**Spanish is cool language**")
          
    #########################################################################################################
    @Rule(Fact(Rooms = 'Yes'), NOT(Fact(Startup = W())))
    def value35(self):
          self.declare(Fact(Startup = input('Do you want to sign up as a startup?')))
          
    @Rule(Fact(Rooms = 'Yes'), NOT(Fact(Person = W())))
    def value36(self):
          self.declare(Fact(Person = input('Do you want to sign up as a person?')))
          
    ##################################################################################################
    @Rule(Fact(Rooms = 'Yes'), Fact(Startup = 'Yes'), NOT(Fact(Table = W())))
    def value37(self):
          self.declare(Fact(Table = input('Do you want to reserve Table?')))
    
    @Rule(Fact(Rooms = 'Yes'), Fact(Startup = 'Yes'), NOT(Fact(Room = W())))
    def value38(self):
          self.declare(Fact(Room = input('Do you want to reserve Room?')))
    ###############################################################################################
    @Rule(Fact(Rooms = 'Yes'), Fact(Startup = 'Yes'), Fact(Table = 'Yes'))
    def value39(self):
          print('**Room 5 contains our best Tables for you**')
    
    @Rule(Fact(Rooms = 'Yes'), Fact(Startup = 'Yes'), Fact(Room = 'Yes'))
    def value40(self):
          print('**Room 4 and Room 6 are created to suit your work**')
          
    ################################################################################################
    @Rule(Fact(Rooms = 'Yes'), Fact(Person = 'Yes'), NOT(Fact(Meeting = W())))
    def value41(self):
          self.declare(Fact(Meeting = input('Do you want to reserve a meeting room?')))
     
    @Rule(Fact(Rooms = 'Yes'), Fact(Person = 'Yes'), NOT(Fact(Work = W())))
    def value42(self):
          self.declare(Fact(Work = input('Do you want to reserve a Work room?')))
     
    @Rule(Fact(Rooms = 'Yes'), Fact(Person = 'Yes'), NOT(Fact(Relaxation = W())))
    def value43(self):
          self.declare(Fact(Relaxation = input('Do you want to relax?')))
          
    #################################################################################################
    @Rule(Fact(Rooms = 'Yes'), Fact(Person = 'Yes'), Fact(Relaxation = 'Yes'))
    def value44(self):
          print('**You can relax in our Cafe or Relaxation room**')
          
    @Rule(Fact(Rooms = 'Yes'), Fact(Person = 'Yes'), Fact(Meeting = 'Yes'), NOT(Fact(one2one = W())))
    def value45(self):
          self.declare(Fact(one2one = input('Do you want to have a 1 to 1 meeting?')))
          
    @Rule(Fact(Rooms = 'Yes'), Fact(Person = 'Yes'), Fact(Meeting = 'Yes'), NOT(Fact(Group = W())))
    def value46(self):
          self.declare(Fact(Group = input('Do you want to have a Group meeting?')))
          
    ####################################################################################################
    @Rule(Fact(Rooms = 'Yes'), Fact(Person = 'Yes'), Fact(Meeting = 'Yes'), Fact(one2one = 'Yes'))
    def value47(self):
          print('**Room 6 is great for one to one meeting**')
          
    @Rule(Fact(Rooms = 'Yes'), Fact(Person = 'Yes'), Fact(Meeting = 'Yes'), Fact(Group = 'Yes'))
    def value48(self):
          print('**Room 5 and Room 6 contains all you need for your team**')
          
    ###################################################################################################
    @Rule(Fact(Rooms = 'Yes'), Fact(Person = 'Yes'), Fact(Work = 'Yes'), NOT(Fact(Private = W())))
    def value49(self):
          self.declare(Fact(Private = input('Do you want Private room?')))
          
    @Rule(Fact(Rooms = 'Yes'), Fact(Person = 'Yes'), Fact(Work = 'Yes'), NOT(Fact(Shared_Room = W())))
    def value50(self):
          self.declare(Fact(Shared_Room = input('Do you want Shared Room room?')))
          
    #######################################################################################################
    @Rule(Fact(Rooms = 'Yes'), Fact(Person = 'Yes'), Fact(Work = 'Yes'), Fact(Private = 'Yes'))
    def value51(self):
        print('Room 3 is great to relax while working in private environment')
          
    @Rule(Fact(Rooms = 'Yes'), Fact(Person = 'Yes'), Fact(Work = 'Yes'), (Fact(Shared_Room = W())))
    def value52(self):
        print('Room 1 and Room 2 have been created to work together on shared rooms')
          
    
    
```

```python
engine = CoworkingSpace()
while(1):
    engine.reset()
    engine.run()
    print('Do you want to run app again?')
    if input() == "No":
        break
```
