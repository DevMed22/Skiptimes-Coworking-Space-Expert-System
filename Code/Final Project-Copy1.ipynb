```python
from experta import *
from tkinter import *
```

```python
class CoworkingSpace(KnowledgeEngine):
    @DefFacts()
    def _intialize(self):
        gui_form1 = Tk()
        gui_form1.title("SkipTimes Coworking Space Expert System")
#         gui_form1.geometry("800x500")
        lb0 = Label(gui_form1,text="Hello now you can book a Course or Room \n let's try",bg="#123456", fg="#fff",width="800",font=('Times',18,'bold'))
        lb0.pack()
        
###############################################################################
        lb1 = Label(gui_form1,text="Do you want to book a Course?",font=('Times',14))
        en1 = Entry(gui_form1,width="40")
        lb1.pack()
        en1.pack()
###############################################################################
        lb2 = Label(gui_form1,text="Do you want book a Room?",font=('Times',14))
        en2 = Entry(gui_form1,width="40")
        lb2.pack()
        en2.pack()
###############################################################################
        lb3 = Label(gui_form1,text="Do you want to sign up as a startup?",font=('Times',14))
        en3 = Entry(gui_form1,width="40")
        lb3.pack()
        en3.pack()
###############################################################################
        lb6 = Label(gui_form1,text="Do you want to sign up as a person?",font=('Times',14))
        en6 = Entry(gui_form1,width="40")
        lb6.pack()
        en6.pack()
##############################################################################
        lb7 = Label(gui_form1,text="Do you want to reserve a meeting room?",font=('Times',14))
        en7 = Entry(gui_form1,width="40")
        lb7.pack()
        en7.pack()

###############################################################################
        lb10 = Label(gui_form1,text="Do you want to relax?",font=('Times',14))
        en10 = Entry(gui_form1,width="40")
        lb10.pack()
        en10.pack()
###############################################################################
        lb11 = Label(gui_form1,text="Do you want to reserve a Work room?",font=('Times',14))
        en11 = Entry(gui_form1,width="40")
        lb11.pack()
        en11.pack()
###############################################################################
        lb14 = Label(gui_form1,text="Do you want programming course?",font=('Times',14))
        en14 = Entry(gui_form1,width="40")
        lb14.pack()
        en14.pack()
###############################################################################
        lb15 = Label(gui_form1,text="Do you want Mobile course?",font=('Times',14))
        en15 = Entry(gui_form1,width="40")
        lb15.pack()
        en15.pack()
###############################################################################
        lb16 = Label(gui_form1,text="Do you want Data Science course?",font=('Times',14))
        en16 = Entry(gui_form1,width="40")
        lb16.pack()
        en16.pack()
###############################################################################
        lb17 = Label(gui_form1,text="Do you want Web course?",font=('Times',14))
        en17 = Entry(gui_form1,width="40")
        lb17.pack()
        en17.pack()
###############################################################################       
        def get_values():
            #lb2["text"] = en2.get()
            self.declare(Fact(Course=en1.get()))
            self.declare(Fact(Rooms=en2.get()))
            self.declare(Fact(Startup=en3.get()))
            self.declare(Fact(Person=en6.get()))
            self.declare(Fact(Meeting=en7.get()))
            self.declare(Fact(Relaxation=en10.get()))
            self.declare(Fact(Work=en11.get()))
            self.declare(Fact(Programming=en14.get()))
            self.declare(Fact(Mobile=en15.get()))
            self.declare(Fact(Data_Science=en16.get()))
            self.declare(Fact(Web=en17.get()))
            
#############################################################################
        btn = Button(gui_form1,command=get_values,text="Submit",bg="#123400", fg="#fff",font=('Times',18))
        btn.pack()
        
        
        gui_form1.mainloop()
        yield Fact(action = "find_book")

###############################################################################################

    @Rule(Fact(Course = 'yes'), Fact(Programming = 'yes'), Fact(Web = 'yes'))
    def value11(self):
        self.declare(Fact(res='React for front, Dijango or node for back'))

    #######################################################################################################################
    @Rule(Fact(Course = 'yes'), Fact(Programming = 'yes'), Fact(Mobile = 'yes'))
    def value13(self):
        self.declare(Fact(res='Flutter'))

    #############
    @Rule(Fact(Course = 'yes'), Fact(Programming = 'yes'), Fact(Mobile = 'yes'))
    def value17(self):
        self.declare(Fact(res='Kotlin'))

    @Rule(Fact(Course = 'yes'), Fact(Programming = 'yes'), Fact(Data_Science = 'yes'))
    def value21(self):
        self.declare(Fact(res='Build models using TensorFlow and Pytorch'))


    ###############################################################################################
    @Rule(Fact(Rooms = 'yes'), Fact(Startup = 'yes'))
    def value39(self):
        self.declare(Fact(res='**Room 5 is good for startup work**'))

    #################################################################################################
    @Rule(Fact(Rooms = 'yes'), Fact(Person = 'yes'), Fact(Relaxation = 'yes'))
    def value44(self):
        self.declare(Fact(res='**You can relax in our Cafe or Relaxation room**'))

    ####################################################################################################
    @Rule(Fact(Rooms = 'yes'), Fact(Person = 'yes'), Fact(Meeting = 'yes'))
    def value47(self):
        self.declare(Fact(res='**Room 6 is great for meeting**'))

    #######################################################################################################
    @Rule(Fact(Rooms = 'yes'), Fact(Person = 'yes'), Fact(Work = 'yes'))
    def value51(self):
        self.declare(Fact(res='Room 3 is great to relax while working'))

    @Rule(Fact(action='find_book'),Fact(res=MATCH.res))
    def res(self, res):
        print("")
        id_res = res
        gui_form2 = Tk()
        gui_form2.title("SkipTimes Coworking Space Expert System")
        gui_form2.geometry("800x500")
        lb0 = Label(gui_form2,text="Your choice",bg="#123456", fg="#fff",width="800",font=('Times',14, 'bold'))
        lb0.pack()
        var = StringVar()
        var.set(id_res)
        lb1 = Label(gui_form2,textvariable = var,bg="#678954", fg="#fff",width="800",font=('Times',14))
        lb1.pack()
        print("Your selection %s\n" %(id_res))
        gui_form2.mainloop()
        
```

```python
engine = CoworkingSpace()
while(1):
    engine.reset()  # Prepare the engine for the execution.
    engine.run()  # Run it!
    print("Do you want to run app again?")
    if input() == "no":
        break;
```
